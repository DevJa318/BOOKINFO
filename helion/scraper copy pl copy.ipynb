```python
import requests
```

```python
import bs4
```

```python
searchlink='https://helion.pl/ksiazki/podstawy-matematyki-w-data-science-algebra-liniowa-rachunek-prawdopodobienstwa-i-statystyka-thomas-nield,pomads.htm#format/d'
```

```python
res = requests.get(searchlink)
```

```python
soup = bs4.BeautifulSoup(res.text)
```
